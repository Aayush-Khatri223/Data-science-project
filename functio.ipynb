```python
def define(*args,**kwargs):
    print("attributs:")
    for arg in args:
        print(f"- {arg}")
    print("details:")
    for keys, value in kwargs.items():
        print(f"{keys}:{value}")
define("aayush",20,"tall","blue eyes",job="engineer")        
```
