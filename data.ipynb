```python
import pandas as pd
import matplotlib.pyplot as plt

# Given data
x = [1, 2, 3, 4, 5]
y = [3, 4, 5, 6, 7]
m = len(x)

alpha = 0.01
theta1 = 0.8

theta_history = []
loss_history = []
iteration_tables = []

# Gradient Descent
for iteration in range(50):
    y_cap = [2 + theta1 * xi for xi in x]
    
    errors = [yi - yci for yi, yci in zip(y, y_cap)]
    sq_errors = [(e)**2 for e in errors]
    x_errors = [xi * e for xi, e in zip(x, errors)]
    
    loss = sum(sq_errors) / (2 * m)
    gradient = -sum(x_errors) / m
    
    theta1 = theta1 - alpha * gradient
    
    theta_history.append(theta1)
    loss_history.append(loss)
    
    # Store iteration table
    df = pd.DataFrame({
        "x": x,
        "y": y,
        "ycap": y_cap,
        "y-ycap": errors,
        "(y-ycap)^2": sq_errors,
        "x(y-ycap)": x_errors
    })
    
    df["iteration"] = iteration + 1
    iteration_tables.append(df)

# -------------------------------
# Show last 5 iterations table
# -------------------------------
last_5 = pd.concat(iteration_tables[-5:], ignore_index=True)
print("\nTABLE FOR LAST 5 ITERATIONS:\n")
print(last_5)

# -------------------------------
# Plot Theta vs Loss
# -------------------------------
plt.figure()
plt.plot(theta_history, loss_history, marker='o')
plt.xlabel("Theta1")
plt.ylabel("Loss J(Theta1)")
plt.title("Gradient Descent: Theta1 vs Loss")
plt.grid()
plt.show()
```
